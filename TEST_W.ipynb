```python
import cPickle
import gzip
import os
import sys
import timeit

import numpy as np
import matplotlib.pyplot as plt

import theano
import theano.tensor as T
from theano.tensor.signal import downsample
from theano.tensor.nnet import conv

from collections import OrderedDict
from collections import defaultdict


def dropout(rng, x, p):
    """ Zero-out random values in x with probability p using rng """
    if p > 0. and p < 1.:
        seed = rng.randint(2 ** 30)
        srng = theano.tensor.shared_randomstreams.RandomStreams(seed)
        mask = srng.binomial(n=1, p=p, size=x.shape,
                dtype=theano.config.floatX)
        return x * mask
    return x

def maxx(x,y):
    return T.switch(x<y, y, x)

class LogisticRegression(object):
    """Multi-class Logistic Regression Class
    The logistic regression is fully described by a weight matrix :math:`W`
    and bias vector :math:`b`. Classification is done by projecting data
    points onto a set of hyperplanes, the distance to which is used to
    determine a class membership probability.
    """

    def __init__(self, input, input_dropout, n_in, n_out):
        """ Initialize the parameters of the logistic regression

        :type input: theano.tensor.TensorType
        :param input: symbolic variable that describes the input of the
                      architecture (one minibatch)

        :type n_in: int
        :param n_in: number of input units, the dimension of the space in
                     which the datapoints lie

        :type n_out: int
        :param n_out: number of output units, the dimension of the space in
                      which the labels lie

        """
        # start-snippet-1
        # initialize with 0 the weights W as a matrix of shape (n_in, n_out)
        self.W = theano.shared(
            value=np.zeros(
                (n_in, n_out),
                dtype=theano.config.floatX
            ),
            name='W',
            borrow=True
        )
        # initialize the baises b as a vector of n_out 0s
        self.b = theano.shared(
            value=np.zeros(
                (n_out,),
                dtype=theano.config.floatX
            ),
            name='b',
            borrow=True
        )

        # symbolic expression for computing the matrix of class-membership
        # probabilities
        # Where:
        # W is a matrix where column-k represent the separation hyper plain for
        # class-k
        # x is a matrix where row-j  represents input training sample-j
        # b is a vector where element-k represent the free parameter of hyper
        # plain-k
        self.p_y_given_x = T.nnet.softmax(T.dot(input, self.W) + self.b)
        self.p_y_given_x_dropout = T.nnet.softmax(T.dot(input_dropout, self.W) + self.b)
        
        # symbolic description of how to compute prediction as class whose
        # probability is maximal
        self.y_pred = T.argmax(self.p_y_given_x, axis=1)
        # end-snippet-1

        # parameters of the model
        self.params = [self.W, self.b]

    def negative_log_likelihood(self, y):
        """Return the mean of the negative log-likelihood of the prediction
        of this model under a given target distribution.

        .. math::

            \frac{1}{|\mathcal{D}|} \mathcal{L} (\theta=\{W,b\}, \mathcal{D}) =
            \frac{1}{|\mathcal{D}|} \sum_{i=0}^{|\mathcal{D}|}
                \log(P(Y=y^{(i)}|x^{(i)}, W,b)) \\
            \ell (\theta=\{W,b\}, \mathcal{D})

        :type y: theano.tensor.TensorType
        :param y: corresponds to a vector that gives for each example the
                  correct label


        """
        # start-snippet-2
        # y.shape[0] is (symbolically) the number of rows in y, i.e.,
        # number of examples (call it n) in the minibatch
        # T.arange(y.shape[0]) is a symbolic vector which will contain
        # [0,1,2,... n-1] T.log(self.p_y_given_x) is a matrix of
        # Log-Probabilities (call it LP) with one row per example and
        # one column per class LP[T.arange(y.shape[0]),y] is a vector
        # v containing [LP[0,y[0]], LP[1,y[1]], LP[2,y[2]], ...,
        # LP[n-1,y[n-1]]] and T.mean(LP[T.arange(y.shape[0]),y]) is
        # the mean (across minibatch examples) of the elements in v,
        # i.e., the mean log-likelihood across the minibatch.
        return -T.mean(T.log(self.p_y_given_x_dropout)[T.arange(y.shape[0]), y])
        # end-snippet-2
    def cost0(self, y):
        return -T.mean(T.log(self.p_y_given_x)[T.arange(y.shape[0]), y])

    def errors(self, y):
        """Return a float representing the number of errors in the minibatch
        over the total number of examples of the minibatch ; zero one
        loss over the size of the minibatch

        :type y: theano.tensor.TensorType
        :param y: corresponds to a vector that gives for each example the
                  correct label
        """

        # check if y has same dimension of y_pred
        if y.ndim != self.y_pred.ndim:
            raise TypeError(
                'y should have the same shape as self.y_pred',
                ('y', y.type, 'y_pred', self.y_pred.type)
            )
        # check if y is of the correct datatype
        if y.dtype.startswith('int'):
            # the T.neq operator returns a vector of 0s and 1s, where 1
            # represents a mistake in prediction
            return T.mean(T.neq(self.y_pred, y))
        else:
            raise NotImplementedError()

class ConvReluPoolLayer(object):
    """ Pool Layer of a convolutional network with Relu activation """

    def __init__(self, rng1, rng2, input, input_dropout, filter_shape, image_shape, poolsize, p_dropout):
        """
        Allocate a ConvReluPoolLayer with shared variable internal parameters.

        :type rng: np.random.RandomState
        :param rng: a random number generator used to initialize weights

        :type input: theano.tensor.dtensor4
        :param input: symbolic image tensor, of shape image_shape

        :type filter_shape: tuple or list of length 4
        :param filter_shape: (number of filters, num input feature maps,
                              filter height, filter width)

        :type image_shape: tuple or list of length 4
        :param image_shape: (batch size, num input feature maps,
                             image height, image width)

        :type poolsize: tuple or list of length 2
        :param poolsize: the downsampling (pooling) factor (#rows, #cols)
        """

        assert image_shape[1] == filter_shape[1]
        self.input = input
        self.input_dropout = input_dropout
        # there are "num input feature maps * filter height * filter width"
        # inputs to each hidden unit
        fan_in = np.prod(filter_shape[1:])
        # each unit in the lower layer receives a gradient from:
        # "num output feature maps * filter height * filter width" /
        #   pooling size
        fan_out = (filter_shape[0] * np.prod(filter_shape[2:]) /
                   np.prod(poolsize))
        # initialize weights with random weights
        W_bound = np.sqrt(6. / (fan_in + fan_out))
        self.W1 = theano.shared(
            np.asarray(
                rng1.uniform(low=-W_bound, high=W_bound, size=filter_shape),
                dtype=theano.config.floatX
            ),
            borrow=True
        )
        self.W2 = theano.shared(
            np.asarray(
                rng2.uniform(low=-W_bound, high=W_bound, size=filter_shape),
                dtype=theano.config.floatX
            ),
            borrow=True
        )

        # the bias is a 1D tensor -- one bias per output feature map
        b_values = np.zeros((filter_shape[0],), dtype=theano.config.floatX)
        self.b1 = theano.shared(value=b_values, borrow=True)
        self.b2 = theano.shared(value=b_values, borrow=True)


        # convolve input feature maps with filters
        conv_out1 = conv.conv2d(
            input=input,
            filters=self.W1,
            filter_shape=filter_shape,
            image_shape=image_shape
        )    
        conv_out2 = conv.conv2d(
            input=input,
            filters=self.W2,
            filter_shape=filter_shape,
            image_shape=image_shape
        ) 
        #zero padding with P=2
        zero = T.zeros((image_shape[0], filter_shape[0], image_shape[2], image_shape[3]), dtype = theano.config.floatX)
        zero_padding1 = T.set_subtensor(zero[:,:,2:image_shape[2]-2,2:image_shape[3]-2], conv_out1)        
        zero_padding2 = T.set_subtensor(zero[:,:,2:image_shape[2]-2,2:image_shape[3]-2], conv_out2)        
        zero_padding = maxx(zero_padding1 + self.b1.dimshuffle('x', 0, 'x', 'x'), zero_padding2 + self.b2.dimshuffle('x', 0, 'x', 'x'))          
        # downsample each feature map individually, using maxpooling
        pooled_out = downsample.max_pool_2d(
            input=zero_padding,
            ds=poolsize,
            ignore_border=True
        )
        # add the bias term. Since the bias is a vector (1D array), we first
        # reshape it to a tensor of shape (1, n_filters, 1, 1). Each bias will
        # thus be broadcasted across mini-batches and feature map
        # width & height   
        self.output = pooled_out * p_dropout

        # convolve input feature maps with filters
        conv_out1 = conv.conv2d(
            input=input_dropout,
            filters=self.W1,
            filter_shape=filter_shape,
            image_shape=image_shape
        )    
        conv_out2 = conv.conv2d(
            input=input_dropout,
            filters=self.W2,
            filter_shape=filter_shape,
            image_shape=image_shape
        ) 
        #zero padding with P=2
        zero = T.zeros((image_shape[0], filter_shape[0], image_shape[2], image_shape[3]), dtype = theano.config.floatX)
        zero_padding1 = T.set_subtensor(zero[:,:,2:image_shape[2]-2,2:image_shape[3]-2], conv_out1)        
        zero_padding2 = T.set_subtensor(zero[:,:,2:image_shape[2]-2,2:image_shape[3]-2], conv_out2)        
        zero_padding = maxx(zero_padding1 + self.b1.dimshuffle('x', 0, 'x', 'x'), zero_padding2 + self.b2.dimshuffle('x', 0, 'x', 'x'))          
        # downsample each feature map individually, using maxpooling
        pooled_out = downsample.max_pool_2d(
            input=zero_padding,
            ds=poolsize,
            ignore_border=True
        )
        # add the bias term. Since the bias is a vector (1D array), we first
        # reshape it to a tensor of shape (1, n_filters, 1, 1). Each bias will
        # thus be broadcasted across mini-batches and feature map
        # width & height   
        self.output_dropout = dropout(rng1,
                                      pooled_out,
                                      p_dropout)
        
        # store parameters of this layer
        self.params = [self.W1, self.b1] + [self.W2, self.b2]
        # keep track of model input
        self.input = input
        self.input_dropout = input_dropout

class HiddenLayer(object):
    def __init__(self, rng1, rng2, input, input_dropout, n_in, n_out, p_dropout, W=None, b=None):
        """
        Typical hidden layer of a MLP: units are fully-connected and have
        sigmoidal activation function. Weight matrix W is of shape (n_in,n_out)
        and the bias vector b is of shape (n_out,).

        NOTE : The nonlinearity used here is tanh

        Hidden unit activation is given by: tanh(dot(input,W) + b)

        :type rng: np.random.RandomState
        :param rng: a random number generator used to initialize weights

        :type input: theano.tensor.dmatrix
        :param input: a symbolic tensor of shape (n_examples, n_in)

        :type n_in: int
        :param n_in: dimensionality of input

        :type n_out: int
        :param n_out: number of hidden units

        :type activation: theano.Op or function
        :param activation: Non linearity to be applied in the hidden
                           layer
        """
        self.input = input
        self.input_dropout = input_dropout
        # end-snippet-1

        # `W` is initialized with `W_values` which is uniformely sampled
        # from sqrt(-6./(n_in+n_hidden)) and sqrt(6./(n_in+n_hidden))
        # for tanh activation function
        # the output of uniform if converted using asarray to dtype
        # theano.config.floatX so that the code is runable on GPU
        # Note : optimal initialization of weights is dependent on the
        #        activation function used (among other things).
        #        For example, results presented in [Xavier10] suggest that you
        #        should use 4 times larger initial weights for sigmoid
        #        compared to tanh
        #        We have no info for other function, so we use the same as
        #        tanh.
        if W is None:
            W_values = np.asarray(
                rng1.uniform(
                    low=-np.sqrt(6. / (n_in + n_out)),
                    high=np.sqrt(6. / (n_in + n_out)),
                    size=(n_in, n_out)
                ),
                dtype=theano.config.floatX
            )

            W1 = theano.shared(value=W_values, name='W1', borrow=True)
            
            W_values = np.asarray(
                rng2.uniform(
                    low=-np.sqrt(6. / (n_in + n_out)),
                    high=np.sqrt(6. / (n_in + n_out)),
                    size=(n_in, n_out)
                ),
                dtype=theano.config.floatX
            )
            W2 = theano.shared(value=W_values, name='W2', borrow=True)

        if b is None:
            b_values = np.zeros((n_out,), dtype=theano.config.floatX)
            b1 = theano.shared(value=b_values, name='b1', borrow=True)
            b2 = theano.shared(value=b_values, name='b2', borrow=True)

        self.W1 = W1
        self.W2 = W2
        self.b1 = b1
        self.b2 = b2

        lin_output1 = T.dot(input, self.W1) + self.b2
        lin_output2 = T.dot(input, self.W2) + self.b2
        self.output = maxx(lin_output1, lin_output2) * p_dropout
        
        lin_output1 = T.dot(input_dropout, self.W1) + self.b2
        lin_output2 = T.dot(input_dropout, self.W2) + self.b2
        lin_dropout = maxx(lin_output1, lin_output2)
        self.output_dropout = dropout(rng2,
                                      lin_dropout,
                                      p_dropout)
        # parameters of the model
        self.params = [self.W1, self.b1] + [self.W2, self.b2]
        
def evaluate_ConvNet(n_epochs, nkerns, batch_size, l2_decay, p_dropouts):
    """ Demonstrates ConvNet on CIFAR-10 dataset

    :type n_epochs: int
    :param n_epochs: maximal number of epochs to run the optimizer

    :type nkerns: list of ints
    :param nkerns: number of kernels on each layer
    """

    rng1 = np.random.RandomState(23455)
    rng2 = np.random.RandomState(23115)


    # allocate symbolic variables for the data
    index = T.lscalar()  # index to a [mini]batch

    # start-snippet-1
    x = T.matrix('x')   # the data is presented as rasterized images
    y = T.ivector('y')  # the labels are presented as 1D vector of
                        # [int] labels

    ######################
    # BUILD ACTUAL MODEL #
    ######################
    print '... building the model'

    # Reshape matrix of rasterized images of shape (batch_size, 3, 32 * 32)
    # to a 4D tensor, compatible with our ConvReluPoolLayer
    # (28, 28) is the size of CIFAR-10 images.
    layer0_input = x.reshape((batch_size, 3, 32, 32)) * p_dropouts[0]
    layer0_input_dropout = dropout(rng1,
                                   x.reshape((batch_size, 3, 32, 32)),
                                   p=p_dropouts[0])
                                   
    # Construct the first convolutional pooling layer:
    # filtering reduces the image size to (32-5+1+4 , 32-5+1+4) = (32, 32)
    # maxpooling reduces this further to (32/2, 32/2) = (16, 16)
    # 4D output tensor is thus of shape (batch_size, nkerns[0], 16, 16)
    layer0 = ConvReluPoolLayer(
        rng1,
        rng2,
        input=layer0_input,
        input_dropout=layer0_input_dropout,
        image_shape=(batch_size, 3, 32, 32),
        filter_shape=(nkerns[0], 3, 5, 5),
        poolsize=(2, 2),
        p_dropout=p_dropouts[1]
    )

    # Construct the second convolutional pooling layer
    # filtering reduces the image size to (16-5+1+4, 16-5+1+4) = (16, 16)
    # maxpooling reduces this further to (16/2, 16/2) = (8, 8)
    # 4D output tensor is thus of shape (batch_size, nkerns[1], 8, 8)
    layer1 = ConvReluPoolLayer(
        rng1,
        rng2,
        input=layer0.output,
        input_dropout=layer0.output_dropout,
        image_shape=(batch_size, nkerns[0], 16, 16),
        filter_shape=(nkerns[1], nkerns[0], 5, 5),
        poolsize=(2, 2),
        p_dropout=p_dropouts[2]
    )
    
    # Construct the second convolutional pooling layer
    # filtering reduces the image size to (8-5+1+4, 8-5+1+4) = (8, 8)
    # maxpooling reduces this further to (8/2, 8/2) = (4, 4)
    # 4D output tensor is thus of shape (batch_size, nkerns[2], 4, 4)
    layer3 = ConvReluPoolLayer(
        rng1,
        rng2,
        input=layer1.output,
        input_dropout=layer1.output_dropout,
        image_shape=(batch_size, nkerns[1], 8, 8),
        filter_shape=(nkerns[2], nkerns[1], 5, 5),
        poolsize=(2, 2),
        p_dropout=p_dropouts[3]
    )

    # classify the values of the fully-connected sigmoidal layer
    layer_hidden_input = layer3.output.flatten(2)
    layer_hidden_input_dropout = layer3.output_dropout.flatten(2)
    
    layer_hidden = HiddenLayer(
        rng1,
        rng2,
        input=layer_hidden_input,
        input_dropout=layer_hidden_input_dropout,
        n_in=nkerns[2] * 4 * 4,
        n_out=nkerns[3],
        p_dropout=p_dropouts[4]
    )
    
    layer_regression = LogisticRegression(input=layer_hidden.output, input_dropout=layer_hidden.output_dropout,
                                          n_in=nkerns[3], n_out=2)
    

    # the cost we minimize during training is the NLL of the model
    cost = layer_regression.negative_log_likelihood(y)
    cost0 = layer_regression.cost0(y)

    # create a list of all model parameters to be fit by gradient descent
    # create a list of gradients for all model parameters
    params = layer_regression.params + layer_hidden.params + layer3.params + layer1.params + layer0.params
    l2_decay /= 2
    for param in params:
        cost += l2_decay * T.sum(param ** 2)
    grads = T.grad(cost, params) 
    
    save_file = open('/afs/cern.ch/user/s/sseifnas/public/save_quark_big7.p')
    for param in params:
        param.set_value(cPickle.load(save_file), borrow=True)
    
    LIST=[]
    for param in params:
        LIST.append(param.get_value(borrow=True))
    
    print 'finish'
    return LIST
   
LIST=evaluate_ConvNet(n_epochs=200, nkerns=[40, 80, 160, 250], batch_size=256, l2_decay=0.0001, p_dropouts=[0.9, 0.75, 0.75, 0.75 , 0.5])
```
